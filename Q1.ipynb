Importing Libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Parameters:

```python
sample_sizes = [2, 10, 30, 150]    # sample sizes
p = 10000   # no. sample points
```

Original Distribution:

```python
def plot_original_distribution(dist_func, dist_params, dist_name):
    data = dist_func(*dist_params, size=p)
    plt.figure(figsize=(5,4))
    sns.histplot(data, bins=50, kde=True, color="#B350FF")
    plt.title(f"{dist_name} Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.show()
    return data
```

Generating Means for CLT

```python
def generate_sampling_distribution(dist_func, dist_params, sample_size):
    return [np.mean(dist_func(*dist_params, size=sample_size)) for _ in range(p)]
```

Plotting Mean Values

```python
def plot_sampling_distributions(dist_func, dist_params, sample_sizes):
    plt.figure(figsize=(8, 6))
    
    for i, n in enumerate(sample_sizes, start=1):
        means = generate_sampling_distribution(dist_func, dist_params, n)
        plt.subplot(2, 2, i)
        sns.histplot(means, bins=50, kde=True, color="#FF83D4")
        plt.title(f"Sample Means (n={n})")
        plt.xlabel("Mean Value")
        plt.ylabel("Density")
    
    plt.tight_layout()
    plt.show()
```

Combining above functions:

```python
def demonstrate_clt(dist_name, dist_func, dist_params, sample_sizes):
    # Show original distribution
    plot_original_distribution(dist_func, dist_params, dist_name)
    
    # Show sampling distributions
    plot_sampling_distributions(dist_func, dist_params, sample_sizes)

demonstrate_clt("Uniform", np.random.uniform, (0, 10), sample_sizes)
demonstrate_clt("Exponential", np.random.exponential, (1,), sample_sizes)
```
